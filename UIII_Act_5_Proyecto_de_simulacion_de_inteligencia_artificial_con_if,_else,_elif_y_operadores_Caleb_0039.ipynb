```python
# --- SIMULACIÓN DE CAJERO INTELIGENTE (IA) ---

# Configuración inicial de datos (Simulando una base de datos)
nombre_cliente = "Alex"
saldo_actual = 5000.0  # [1]
pin_secreto = 1234  # [1]

print("IA: ¡Hola! Soy tu asistente bancario virtual. Detecto tu presencia.")
print("IA: Por favor, para continuar, identifícate.")

# Entrada interactiva del usuario
pin_ingresado = int(input("CLIENTE: Ingresa tu PIN de 4 dígitos: ")) # [2]

# Estructura if-else para autenticación con operador relacional == [3, 4]
if pin_ingresado == pin_secreto:
    print(f"IA: Acceso concedido. Es un gusto verte de nuevo, {nombre_cliente}.")
    print("IA: ¿Qué operación deseas realizar hoy?")
    print("Opciones: 1. Consultar Saldo | 2. Retirar Dinero | 3. Depositar")

    opcion = input("CLIENTE (Elige 1, 2 o 3): ") # [2]

    # Estructura elif para toma de decisiones múltiple [3, 5]
    if opcion == "1":
        # Relacional: Verifica si el cliente tiene saldo positivo [4]
        if saldo_actual > 0:
            print(f"IA: Tu saldo actual es de ${saldo_actual}. ¡Tienes fondos suficientes!")
        else:
            print(f"IA: Tu cuenta está en cero. ¿Te gustaría realizar un depósito?")

    elif opcion == "2":
        monto_retiro = float(input("IA: ¿Cuánto dinero deseas retirar?: "))

        # Operador lógico 'and' y relacionales para validar la operación [4, 6]
        if monto_retiro > 0 and monto_retiro <= saldo_actual:
            # Operador aritmético de resta para actualizar saldo [7, 8]
            saldo_actual = saldo_actual - monto_retiro
            print(f"IA: Procesando retiro de ${monto_retiro}...")
            print(f"IA: ¡Éxito! Tu nuevo saldo es ${saldo_actual}. No olvides tu dinero.")
        elif monto_retiro > saldo_actual:
            print("IA: Error. El monto solicitado excede tu saldo disponible.")
        else:
            print("IA: Lo siento, no puedo procesar montos negativos o nulos.")

    elif opcion == "3":
        monto_deposito = float(input("IA: ¿Qué cantidad vas a depositar?: "))

        # Validación con if simple [3]
        if monto_deposito > 0:
            # Operador aritmético de suma [7, 8]
            saldo_actual = saldo_actual + monto_deposito
            print(f"IA: Depósito recibido. Ahora tienes ${saldo_actual} en tu cuenta.")
        else:
            print("IA: La cantidad introducida no es válida para un depósito.")

    else:
        print("IA: No reconozco esa opción. Por favor, intenta de nuevo más tarde.")

else:
    # Caso en que el PIN sea incorrecto usando operador relacional != implícito [4]
    print("IA: Alerta de seguridad. El PIN es incorrecto. El sistema se cerrará.")

print("IA: Gracias por usar nuestros servicios. ¡Que tengas un excelente día!")
```
